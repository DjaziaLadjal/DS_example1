## Scaling the data to population size

For a same unit of time, number of tweets in the USA will be higher than the number of tweets in the UK, this is because the population in the USA is 5 times higher. To be able to compare twitter activities between to 2 countries, it is important to scale to the population size. In this case, we can devide the number of tweets by the pupulation number in millions and that would give us 'number of tweets by 1 million people'.
We will add a scaled attribute to the dataframe which will represent the number of tweets by 1 million people.
Replace in the code below the vcalue 1. by the population number in the UK.

```python
twitter_counts['count_scaled']=twitter_counts['count']/1.
```

If you scaled the data correctly you should see a difference when plotting the count and scaled_count together

```python
fig, ax = plt.subplots()
ax.plot(twitter_counts['day'],twitter_counts['count'],color='blue')
ax.plot(twitter_counts['day'],twitter_counts['count_scaled'],color='red')
ax.set(xlabel='day', ylabel='number of tweets',
       title='Twitter daily counts',xlim=[0,250])

ax.grid()
plt.show()
```

From now on we will be using the scaled_counts for the rest of the pre-processing

We can smooth the data using a rolling average. Go ahead and change the window size of the rolling average to smooth your data.

```python
twitter_counts['rolling_average'] = twitter_counts['count_scaled'].rolling(window=1).mean()
```

If you changed the window size for the rolling average you should see a difference between the count_scaled and the averaged count.

```python
fig, ax = plt.subplots()
ax.plot(twitter_counts['day'],twitter_counts['count_scaled'],color='blue')
ax.plot(twitter_counts['day'],twitter_counts['rolling_average'],color='red')
ax.set(xlabel='day', ylabel='number of tweets',
       title='Twitter daily counts',xlim=[0,250])

ax.grid()
plt.show()
```

```python
regression=LinearRegression().fit(twitter_counts[twitter_counts['rolling_average'].notnull()][['day']],\
                                  twitter_counts[twitter_counts['rolling_average'].notnull()]['rolling_average'])
#regression.score(twitter_counts[['day']],twitter_counts['rolling_average'])
```

```python
twitter_counts['baseline'] = regression.predict(twitter_counts[['day']])
```

```python
fig, ax = plt.subplots()
ax.plot(twitter_counts['day'],twitter_counts['count_scaled'],color='blue')
ax.plot(twitter_counts['day'],twitter_counts['baseline'],color='red')
ax.set(xlabel='day', ylabel='number of tweets',
       title='Twitter daily counts',xlim=[0,250])

ax.grid()
plt.show()
```

```python
twitter_counts['events']=twitter_counts['count_scaled']-twitter_counts['baseline']
```

```python
fig, ax = plt.subplots()
ax.plot(twitter_counts['day'],twitter_counts['count_scaled'],color='blue')
ax.plot(twitter_counts['day'],twitter_counts['baseline'],color='red')
ax.plot(twitter_counts['day'],twitter_counts['events'],color='green')
ax.set(xlabel='day', ylabel='number of tweets',
       title='Twitter daily counts',xlim=[0,250])

ax.grid()
plt.show()
```

```python
twitter_counts['residual']=twitter_counts['count_scaled']-twitter_counts['events']
```

```python
fig, ax = plt.subplots()
ax.plot(twitter_counts['day'],twitter_counts['count_scaled'],color='blue')
ax.plot(twitter_counts['day'],twitter_counts['baseline'],color='red')
ax.plot(twitter_counts['day'],twitter_counts['residual'],color='green')
ax.set(xlabel='day', ylabel='number of tweets',
       title='Twitter daily counts',xlim=[0,250])

ax.grid()
plt.show()
```
